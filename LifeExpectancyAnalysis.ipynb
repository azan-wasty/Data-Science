```python
for column in df.columns:
    if column != 'Status':
        if df[column].isnull().any():
            median_value = df[column].median()
            df[column].fillna(median_value, inplace=True)

print(df.isnull().sum())
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
from scipy import stats
import numpy as np

if not df.empty:

    numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    if 'Status' in numerical_cols:
        numerical_cols.remove('Status')

    print(f"Analyzing numerical columns: {numerical_cols}\n")

    # Method 1: Z-Score Method

    print("Z-SCORE METHOD ")


    z_scores = np.abs(stats.zscore(df[numerical_cols]))
    outliers_z = (z_scores > 3).any(axis=1)

    print(f"\nTotal outliers found: {outliers_z.sum()}")
    print(f"Percentage of outliers: {(outliers_z.sum() / len(df)) * 100:.2f}%\n")

    if outliers_z.sum() > 0:
        print("Outliers by Z-score:")
        display(df[outliers_z])

    # Method 2: IQR Method

    print("IQR METHOD (threshold: Q1-1.5*IQR, Q3+1.5*IQR)")


    outliers_iqr = pd.Series([False] * len(df), index=df.index)

    for column in numerical_cols:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        col_outliers = (df[column] < lower_bound) | (df[column] > upper_bound)
        outliers_iqr = outliers_iqr | col_outliers

        print(f"\n{column}:")
        print(f"  Q1={Q1:.2f}, Q3={Q3:.2f}, IQR={IQR:.2f}")
        print(f"  Bounds: [{lower_bound:.2f}, {upper_bound:.2f}]")
        print(f"  Outliers: {col_outliers.sum()}")

    print(f"\n\nTotal outliers found: {outliers_iqr.sum()}")
    print(f"Percentage of outliers: {(outliers_iqr.sum() / len(df)) * 100:.2f}%\n")

    if outliers_iqr.sum() > 0:
        print("Outliers by IQR:")
        display(df[outliers_iqr])

    # Comparison

    print("COMPARISON")

    print(f"Outliers detected by Z-score only: {(outliers_z & ~outliers_iqr).sum()}")
    print(f"Outliers detected by IQR only: {(outliers_iqr & ~outliers_z).sum()}")
    print(f"Outliers detected by both methods: {(outliers_z & outliers_iqr).sum()}")

    #  Remove outliers
   #Z Score methos is better in my opinion as it takes all standard deviation , mean , median into account o determine the outliers

    print("DATA CLEANING OPTIONS")

    print(f"Original dataset size: {len(df)}")
    print(f"After removing Z-score outliers: {len(df[~outliers_z])}")
    print(f"After removing IQR outliers: {len(df[~outliers_iqr])}")
    print(f"After removing both: {len(df[~(outliers_z | outliers_iqr)])}")


    df_clean_z = df[~outliers_z]
    df_clean_iqr = df[~outliers_iqr]
    df = df[~(outliers_z | outliers_iqr)]

```

```python
# One-hot encode the 'Status' column
df = pd.get_dummies(df, columns=['Status'], prefix='Status')

# Display the value counts of the new columns and sample data
print(df[['Status_Developing', 'Status_Developed']].value_counts())
print("\nSample data:")
display(df[['Year', 'Life expectancy ', 'Status_Developing', 'Status_Developed']].head(10))
```
