```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("LinearRegression.csv")
x=df['Temperature'].values
y = df['Revenue'].values
display("Independent Variable (x)",x)
display("Dependent Variable (y)",y)
```

```python
x_sq = x**2
n = len(x)
sum_xy = np.sum(x * y)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_x_sq = np.sum(x_sq)

b = (n * sum_xy - sum_x * sum_y) / (n * sum_x_sq - sum_x**2)
a = (sum_y - b * sum_x) / n

print("Slope (b):", b)
print("Intercept (a):", a)
```

```python
def predict(input_x):
    return a + b * input_x
```

```python
y_pred=[0,0]

y_pred[0]=(predict(27))
y_pred[1]=(predict(40))
```

```python
print("\nPrediction:Revenue for 27 and 40 predicted are:")
print (y_pred[0])
print (y_pred[1])
```

```python
mean_y = np.mean(y)
y_pred_scratch = predict(x)
means_squared_error = np.mean((y - y_pred_scratch)**2)
print ("\nMean Squared Error from scratch:", means_squared_error)
ss_residual = np.sum((y - y_pred_scratch)**2)
ss_total = np.sum((y - mean_y)**2)
r_squared = 1 - (ss_residual / ss_total)
print("R-squared from scratch:", r_squared)
```

```python
x_line = np.linspace(min(x) - 1, max(x) + 1, 100)
y_line = predict(x_line)
# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual data points')
plt.plot(x_line, y_line, color='red', label=f'Regression Line (y = {a:.2f} + {b:.2f}x)')
plt.title('Simple Linear Regression from Scratch')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(
x, y, test_size=0.1, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train.reshape(-1, 1), y_train)
y_pred = model.predict(x_test.reshape(-1, 1))
y_pred_custom = [0, 0]
y_pred_custom[0] = model.predict([[27]])
y_pred_custom[1] = model.predict([[40]])
```

```python
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
print("Model Coefficient (Slope):", model.coef_[0])
print("Model Intercept:", model.intercept_)
```

```python
print("\nPrediction:Revenue for 27 and 40 predicted are:")
print (y_pred_custom[0])
print (y_pred_custom[1])
```

```python
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, model.predict(x.reshape(-1, 1)), color='red', label='Regression Line')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```
